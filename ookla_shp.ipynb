```python
plt.figure(figsize=(30,40))
chart = sns.boxplot(x="HEATMAP_MUNICIPALITY", y="DOWNLOAD SPEED", data=df,
    
    palette='Set1'
    
)
chart.set_xticklabels(chart.get_xticklabels(),rotation=90,fontweight='light',
    fontsize='20')
chart.set_xlabel("HEATMAP MUNICIPALITY",fontsize=30)
chart.set_ylabel("DOWNLOAD SPEED",fontsize=30)
plt.ylim(0, 400)
plt.savefig('RMA_boxplot_download.pdf', bbox_inches='tight')
```

```python
df_2020.head()
```

```python
df_2020_his=df_2020[['DOWNLOAD SPEED','HEATMAP_MUNICIPALITY']]
df_2020_his.head()
```

```python
df_2020_his['DOWNLOAD SPEED'].plot(kind='hist', bins=10, figsize=(8, 5))

plt.title('Histogram of DOWNLOAD SPEED from all counties in 2020') # add a title to the histogram
plt.ylabel('Number of test') # add y-label
plt.xlabel('Download speed') # add x-label
plt.savefig('RMA_histogramplot_download_2020.png',dpi=300, bbox_inches='tight', pad_inches=0)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks", color_codes=True)
```

```python
  
    

chart=sns.catplot(x="HEATMAP_MUNICIPALITY", y="DOWNLOAD SPEED", kind="strip", s=5,height=9.27, aspect=13.7/9.27, 
                  data=df_2020_his, palette='Set1')
chart.set_xticklabels(rotation=90)
chart.set(ylim=(0, 100))
plt.savefig('RMA_catplot_download_2020_ylim=100.png',dpi=300, bbox_inches='tight', pad_inches=0)
plt.show()
```
